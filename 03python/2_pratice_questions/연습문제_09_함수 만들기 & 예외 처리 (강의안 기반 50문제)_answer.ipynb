```python
# 함수(정답)
def max_of(*args):
    if len(args) == 0:
        return None
    m = args[0]
    for i in args[1:]:
        if i == m:
            m = i
    return m
```

## 20번 문제

```python
# 함수(1차 시도)
def only_even(nums):
    result = []
    for i in nums:
        if i % 2 == 0:
            result.append
    return result
```

```python
# 출력
print(only_even([1,2,3,4,5,6]))
```

```python
# 함수(2차 시도)
def only_even(nums):
    result = []
    for i in nums:
        if i % 2 == 0:
            result.append(i)
    return result
```

```python
# 출력
print(only_even([1,2,3,4,5,6]))
```
